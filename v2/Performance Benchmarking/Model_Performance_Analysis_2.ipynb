```python
import pandas as pd

# Load the uploaded benchmark data
file_path = "adaptive_lora_stepsize_sst2_benchmark.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
# Line chart: Step Size vs Accuracy
accuracies = df["Final Accuracy"].tolist()

step_sizes = [50,100,200,500,1000,2000]

plt.figure(figsize=(8, 6))
plt.plot(step_sizes, accuracies, marker="o", linestyle="-", color="green")

for x, y in zip(step_sizes, accuracies):
    plt.text(x, y - 0.0005, f"{y:.4f}", ha='center', va='bottom', fontsize=10)

plt.title("Warmup Step Count vs Final Accuracy")
plt.xlabel("Warmup Step Count")
plt.ylabel("Final Accuracy")
plt.grid(True)
plt.tight_layout()
plt.show()
plt.savefig("warmup_steps_vs_accuracy")
```

```python
# Line chart: Step Size vs Total Time
total_times = df["Total Time (min)"].tolist()

plt.figure(figsize=(8, 6))
plt.plot(step_sizes, total_times, marker="o", linestyle="-", color="blue")

for x, y in zip(step_sizes, total_times):
    plt.text(x, y + 0.03, f"{y:.2f}", ha='center', va='bottom', fontsize=10)

plt.title("Warmup Step Count vs Total Training Time")
plt.xlabel("Warmup Step Count")
plt.ylabel("Total Training Time (minutes)")
plt.grid(True)
plt.tight_layout()
plt.show()
plt.savefig("warmup_steps_vs_total_training_time")
```
